#### **What makes Taylor Series special?:**
* Taylor Series enable us to approximately re-express a complex function using a series of simpler functions; this was the short answer, however, if you are looking for the complete answer, I assure that you are in the right place.
At the end of this module, you can apply this amazing approach to any problems, not only by hand in your notebook, but also by programing in a notebook like this one.<br><br>

* Consider the function $f(x)= cos(x)$. If trigonametry makes you feel uneasy or unwelcome to explore, I have to confess that you are not alone at all. I'm going to intruduce you to our savier, Taylor Series whose superhero power is to diappeare the cosine function $cos(Î¸)$ and turn it to a quadratic expression of the form $(ax^2 +bx+c)$ which doesn't even contain any reference to trigonametry.<br><br>

* In fact, it is used to build an approximation to any complex function using simpler functions. It is composed of a sequence of gradually improving approximations. It just needs to know EVERYTHING about a specific point in the complex function, and by everything I mean the function's value, Jacobian, Hessian ... at the point. Let's see it through an example to build up your intuition, so that you can understand this concept geometrically.
<br>

$(0.1)$
$$ zero_{-} degree_{-} polynomial : f_0(x) = c ,$$<br>

----
$(0.2)$
$$ first_{-} degree_{-} polynomial : f_1(x) = bx+c,$$<br>
$$f_1(x) = f^J(x)x+f(x)$$

----
$(0.3)$
$$ second_{-} degree_{-} polynomial : f_2(x) = ax^2+bx+c,$$<br>
$$f_2(x) = \frac{1}{2}f^H(x)x^2+f^J(x)x+c$$
$$....$$

* As you can see, the function of zero degree $(0.1)$ is just a straight horizontal line at the point x. $figure(1.1)$. It is a flat line and not even a function of x, because the only piece of information which is given to it is the function's value at the point x, as a result the approximation is quite poor. We can give it two pieces of information which is the function's gradient at our point of interest x $(0.1)$. Now, our approximation is a straight line but this time with the same gradient as the complex function at the point.$figure(1.1)$ Notice that each term in the series provides a better approximation. The more information you give, the more accurate approximation you get.
<br>

* Let's consider $f(x) = cos(x)$ and apply our new approach to it.
$$f_0(x) = f(0)  â†’ f^0(x) = cos(x)  ,$$$$  f^0(x) = 1 $$
* *First approximation is just a straight horizontal line at $f(x) = 1$. This is a poor approximation, however, there is still room to improve.*
--------
$$f_1(x) = f'(0)x+f(0) ,$$
$$ f'(x) = -sin(x) â†’ f'(0) = 0,$$
$$f_1(x) = (0)x+ 1  $$$$ f_1(x) = 1 $$
* *This term cannot make any improvement because in $x=0$ the gradient is zero.*
------
$$f_2(x) = \frac{1}{2} f''(0)x^2+ f'(0)x+f(0) $$$$ f''(x) = -cos(x) â†’ f''(0) = -1 $$
$$f_2(x) = \frac{-1}{2} x^2 + 1 $$

* *The second degree polymonial which is a reletivly good approximation. Notice that these zeros will occur whenever differentiate an odd number of times.*
-----
<br><br>
* I hope you already have noticed the pattern. We can generilize this pattern to give us the polynomial expression of n degree without working through it as follows:

$(0.4)$
$$ f_n(x) =\sum \limits _{i=0} ^{n}\frac{(-1)^nx^{2n}}{2n!}$$ <br>


$(0.5)$
$$ f_n(x) =1 - \frac{x^2}{2}+\frac{x^4}{6} ....$$
<br>


* I have to confess that doesn't count as a Taylor Series, by the way. When we are specifically looking at the point zero, this case is refered to as a Maclaruin Series. The generalised form of Maclaurin and Taylor series at the point $p$ are as follows:





$(0.6)$
$$ Maclaurin Series Formula : T =\sum \limits _{i=0} ^{n}\frac{f^n(0)}{n!} x^n  $$
<br>
$(0.7)$
$$ Talor Series Formula : T =\sum \limits _{i=0} ^{n}\frac{f^n(p)}{n!} (x-p)^n  $$
<br><br>
* Using this series, you can easily build up an approximation to any complex function which is continues everywhere and well-behaved. Let's now ask our computers to plot whatever we have learned here.


***
$\mathbf{\text{1. Figures: Taylor Series}}$<br>
***

```python
# Importing required libraries
import numpy as np
import matplotlib.pyplot as plt
import math

# Taylor series definition {FIGURE 1.1}
#>>>>>>>>>>>>>>>>>>>>>>>>>
n = 1
ts= lambda x,n: sum(((-1)**i) * (x ** (2*i)) / (math.factorial(2 * i)) for i in range(n+1))
names = ["zero degree expression $f_0(x)$","second degree expression $f_2(x)$"]
#<<<<<<<<<<<<<<<<<<<<<<<<<


#Creating a vector of 100 various values of x
#between -20 and +20 to help us plot our function.
x = np.linspace(-20,+20,100)


for i in range(n+1): #Implementing Sigma (sum) function.

  #Plotting Configuration
  plt.rcParams['figure.figsize'] = [6.0, 3.0]
  plt.rcParams['figure.dpi'] = 80
  plt.plot(x,ts(x,i) , color = 'red')
  plt.plot(x,np.cos(x) , color = 'blue')
  plt.xlim((-6,6))
  plt.ylim((-3,3))
  plt.xlabel(names[i])
  plt.title("Figure 1.1")
  plt.grid(visible=True)
  plt.show()
```

$Practice (2.1):$<br>
* As you saw above, the corresponding Maclaurin series for the cosine fuction was as follow:
<br>

$(0.4)$
$$ f_n(x) =\sum \limits _{i=0} ^{n}\frac{(-1)^nx^{2n}}{2n!}$$ <br>

```python
# Importing required libraries
import numpy as np
import matplotlib.pyplot as plt
import math

# Maclaurin Series definition
#>>>>>>>>>>>>>>>>>>>>>>>>>
n = 8
mc= lambda x,n: sum(((-1)**i) * (x ** (2*i)) / (math.factorial(2 * i)) for i in range(n+1))
#<<<<<<<<<<<<<<<<<<<<<<<<<


#Creating a vector of 100 various values of x
#between -20 and +20 to help us plot our function.
x = np.linspace(-20,+20,100)


for i in range(n+1): #Implementing Sigma (sum) function.

  #Plotting Configuration
  plt.rcParams['figure.figsize'] = [6.0, 3.0]
  plt.rcParams['figure.dpi'] = 80
  plt.plot(x,mc(x,i) , color = 'red')
  plt.plot(x,np.cos(x) , color = 'blue')
  plt.xlim((-6,6))
  plt.ylim((-3,3))
  plt.xlabel("Maclaurin Series up to $f_{0}(x)$".format(i))
  plt.title("Assessment: 2.1")
  plt.grid(visible=True)
  plt.show()
```

$Practice (2.2):$<br>
* The function $f(x) = e^x$ is absolutely well-behaved and continuos everywhere. Its Maclaurin Series can be calulated as follows:
<br>

$$ f_n(x) =1 + x + \frac{x^2}{2!}+\frac{x^3}{3!} ....$$
<br>

$$ f_n(x) =\sum \limits _{i=0} ^{n}\frac{x^{n}}{n!}$$ <br>

* And the corresponding Taylor Series at the point $p$ can also be calulated in a similar way as follows:<br><br>

$$ f_n(x) =\sum \limits _{i=0} ^{n}\frac{e^p(x-p)^{n}}{n!}$$
<br>
<br>
$$ f_n(x) =\frac{e^p}{1}+\frac{e^p (x-p)}{1}+\frac{e^p (x-p)^2}{2} ....$$




```python
# Importing required libraries
import numpy as np
import matplotlib.pyplot as plt
import math

# Maclaurin series definition for f(x) = e^x
#>>>>>>>>>>>>>>>>>>>>>>>>>
n = 8
mc= lambda x,n: sum((x**i)/(np.math.factorial(i)) for i in range(n+1))
#<<<<<<<<<<<<<<<<<<<<<<<<<


#Creating a vector of 100 various values of x
#between -20 and +20 to help us plot our function.
x = np.linspace(-20,+20,100)


for i in range(n): #Implementing Sigma (sum) function.
  pdf = mc(x,i)

  #Plotting Configuration
  plt.rcParams['figure.figsize'] = [6.0, 3.0]
  plt.rcParams['figure.dpi'] = 80
  plt.plot(x,pdf , color = 'red')
  plt.plot(x,np.math.e**x , color = 'blue')
  plt.xlim((-6,6))
  plt.ylim((-3,3))
  plt.xlabel("Maclaurin Series up to $f_{0}(x)$".format(i))
  plt.title("Assessment: 2.2")
  plt.grid(visible=True)
  plt.show()
```

```python
# Importing required libraries
import numpy as np
import matplotlib.pyplot as plt
import math

# Taylor series definition for f(x) = e^x at the point p = 1
#>>>>>>>>>>>>>>>>>>>>>>>>>
n = 8
p = 1
ts= lambda x,n: sum(((np.math.e**p)*(x-p)**i)/(np.math.factorial(i)) for i in range(n+1))
#<<<<<<<<<<<<<<<<<<<<<<<<<


#Creating a vector of 100 various values of x
#between -20 and +20 to help us plot our function.
x = np.linspace(-20,+20,100)


for i in range(n): #Implementing Sigma (sum) function.

  #Plotting Configuration
  plt.rcParams['figure.figsize'] = [6.0, 3.0]
  plt.rcParams['figure.dpi'] = 80
  plt.plot(x,ts(x,i) , color = 'red')
  plt.plot(x,np.math.e**x , color = 'blue')
  plt.xlim((-6,6))
  plt.ylim((-3,3))
  plt.xlabel("Taylor Series up to $f_{0}(x)$".format(i))
  plt.title("Assessment: 2.2")
  plt.grid(visible=True)
  plt.show()
```

##### *For the **multivariable fuction**$ f(x,y) = e^{-(x^2-y^2)}$, the second-degree Taylor polynomial at the point (a,b) is:*
<br>
$$ Talor Series Formula : f_2(x+a,y+b) =f(a,b)+\partial f(a,b) \begin{bmatrix}
x-a\\y-b\end{bmatrix}+\frac{1}{2}\begin{bmatrix}
x-a & y-b\end{bmatrix}Hf(a,b)\begin{bmatrix}
x-a\\y-b\end{bmatrix}$$

<br>
For $a,b = 0$ we have:
$$ f_2(x,y) = 1- x^2 - y^2$$

<br>

```python
# Importing required libraries
import numpy as np
import matplotlib.pyplot as plt

# Taylor series definition for f(x,y) = ð‘’^âˆ’(ð‘¥2âˆ’ð‘¦2)
#>>>>>>>>>>>>>>>>>>>>>>>>>
z = lambda x,y: np.e**(-(x**2+y**2))
mc = lambda x,y: 1 - x**2 - y**2
#<<<<<<<<<<<<<<<<<<<<<<<<<

# defining surface and axes
x = np.outer(np.linspace(-2, 2, 10), np.ones(10))
y = x.copy().T

#Plotting Configuration
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, z(x,y), cmap='viridis',edgecolor='green')
ax.plot_surface(x, y, mc(x,y), cmap='viridis',edgecolor='green')
ax.set_title('Assessment: 2.3')
plt.rcParams['figure.figsize'] = [10, 5]
plt.rcParams['figure.dpi'] = 100
plt.show()
```
